###The Number Guessing Game (CLI)
The Number Guessing Game (CLI) is a simple command-line program where the system generates a random number (1–100) and the user tries to guess it. After each guess, the program gives hints like “too high” or “too low” to guide the user.

It keeps track of the number of attempts and displays it when the correct number is guessed. The program also handles invalid inputs and allows the user to exit anytime.

This project demonstrates basic concepts such as loops, conditionals, random number generation, and input validation.

```python
import random

class GuessGame:
    def __init__(self, low=1, high=100):
        self.low = low
        self.high = high
        self.target = random.randrange(low, high + 1)
        self.tries = 0

    def check_guess(self, value):
        self.tries += 1
        if value == self.target:
            return "correct"
        return "low" if value < self.target else "high"

    def start(self):
        print(f"\nGuess the number between {self.low} and {self.high}")
        print("Enter 'q' to quit\n")

        while True:
            data = input("Your guess: ").strip()

            if data.lower() == 'q':
                print(f"Game stopped. The number was {self.target}")
                break

            if not data.isdigit():
                print("Invalid input. Enter a valid number.")
                continue

            num = int(data)
            result = self.check_guess(num)

            if result == "low":
                print("Too low ↑")
            elif result == "high":
                print("Too high ↓")
            else:
                print("\nYou got it!")
                print(f"Attempts taken: {self.tries}")
                break


def main():
    game = GuessGame()
    game.start()


if __name__ == "__main__":
    main()
```
